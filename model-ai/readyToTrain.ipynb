```python
from ultralytics import YOLO

# load model pre-trained segmentation
model = YOLO("yolo11s-seg.pt")

# train dengan dataset roboflow
model.train(
    data="datasets2/data.yaml",  # path ke data.yaml dari roboflow
    epochs=50,                  # jumlah epoch
    imgsz=640,                  # ukuran input
    device=0,                   # pakai GPU
    batch=16                    # sesuaikan dengan VRAM
)
```

Validasi (evaluate) di split validation

```python
from ultralytics import YOLO

model = YOLO("runs/segment/train4/weights/best.pt")

results = model.val(
    data="datasets2/data.yaml",
    split="val",       # 'val' untuk validation, 'test' untuk test
    imgsz=640,
    batch=16,
    device=0
)

# ----- Overall metrics -----
print("Mask mAP50:     ", results.seg.map50)   # mAP@0.50 untuk mask
print("Mask mAP50-95:  ", results.seg.map)     # mAP@[0.50:0.95] untuk mask
print("Mask Precision: ", results.seg.p.mean())  # rata-rata P semua kelas
print("Mask Recall:    ", results.seg.r.mean())  # rata-rata R semua kelas)

print("Box mAP50:      ", results.box.map50)   # mAP@0.50 untuk bbox
print("Box mAP50-95:   ", results.box.map)     # mAP@[0.50:0.95] untuk bbox
print("Box Precision:  ", results.box.p.mean())
print("Box Recall:     ", results.box.r.mean())

# ----- Per-class AP (optional) -----
# results.seg.ap50 dan results.seg.ap berbentuk array [num_classes]
# urutan kelas mengikuti results.names (dict index->name)
for i, name in results.names.items():
    print(f"[{name}] Mask AP50={results.seg.ap50[i]:.3f}, Mask AP50-95={results.seg.ap[i]:.3f}")
```

Testing

```python
test_res = model.val(
  data="datasets2/data.yaml", 
  split="test", imgsz=640, 
  batch=16, 
  device=0
  )
print("TEST Mask mAP50:", test_res.seg.map50, " | Mask mAP50-95:", test_res.seg.map)
```

```python
from ultralytics import YOLO

# load model pre-trained segmentation dengan model yang lebih besar
model = YOLO("yolov8x-seg.pt")  # menggunakan YOLOv8 extra large untuk performa lebih baik

# train dengan dataset roboflow dan strategi yang lebih agresif
model.train(
    data="datasets2/data.yaml",     # path ke data.yaml dari roboflow
    epochs=200,                     # menambah jumlah epoch significantly
    imgsz=1024,                    # meningkatkan ukuran input untuk detail lebih baik
    device=0,                       # pakai GPU
    batch=8,                       # kurangi batch size untuk stabilitas
    patience=50,                   # early stopping patience
    lr0=0.001,                     # initial learning rate yang lebih kecil
    lrf=0.0001,                    # final learning rate yang lebih kecil
    momentum=0.95,                 # tingkatkan momentum
    weight_decay=0.001,            # tingkatkan weight decay untuk regularisasi
    warmup_epochs=5,              # lebih banyak warmup epochs
    box=10.0,                     # tingkatkan box loss gain
    cls=0.8,                      # sesuaikan cls loss gain
    dfl=2.0,                      # tingkatkan dfl loss gain
    mask_ratio=6,                 # tingkatkan mask loss gain
    overlap_mask=True,            # mask overlap untuk akurasi lebih tinggi
    augment=True,                 # gunakan augmentasi
    mixup=0.3,                    # tambahkan mixup augmentation
    mosaic=1.0,                   # maksimalkan mosaic
    copy_paste=0.4,              # tambahkan copy-paste augmentation
    degrees=45.0,                # rotasi yang lebih agresif
    translate=0.2,               # translasi
    scale=0.5,                   # scaling
    shear=10.0,                  # shearing
    perspective=0.001,           # perspective transformation
    flipud=0.5,                  # vertical flip
    fliplr=0.5,                  # horizontal flip
    hsv_h=0.015,                # HSV augmentation
    hsv_s=0.7,
    hsv_v=0.4,
    hide_labels=False,           # jangan sembunyikan label
    save=True,                   # simpan hasil
    save_period=10               # simpan model setiap 10 epochs
)
```

```python
from ultralytics import YOLO

# load model pre-trained segmentation
model = YOLO("yolo11n.pt")  # menggunakan model yang lebih besar

# train dengan dataset roboflow
model.train(
    data="datasets2/data.yaml",  # path ke data.yaml dari roboflow
    epochs=100,                  # menambah jumlah epoch
    imgsz=640,                  # ukuran input
    device=0,                   # pakai GPU
    batch=16,                   # sesuaikan dengan VRAM
    patience=50,                # early stopping patience
    lr0=0.01,                  # initial learning rate
    lrf=0.001,                 # final learning rate
    momentum=0.937,            # SGD momentum/Adam beta1
    weight_decay=0.0005,       # optimizer weight decay
    warmup_epochs=3,           # warmup epochs
    box=7.5,                   # box loss gain
    cls=0.5,                   # cls loss gain
    dfl=1.5,                   # dfl loss gain
    mask_ratio=4,              # mask loss gain
    overlap_mask=True,         # mask overlap (True = higher accuracy)
    augment=True               # use augmentation
)
```

```python
from ultralytics import YOLO

# Load model hasil training (gunakan best weights)
model = YOLO("runs/detect/train/weights/best.pt")

# Evaluasi pada validation set
results = model.val(
    data="datasets2/data.yaml",
    split="val",
    imgsz=640,
    batch=16,
    device=0
)

print("\n----- Metrics Hasil Training -----")
print(f"Box mAP50:      {results.box.map50:.3f}")
print(f"Box mAP50-95:   {results.box.map:.3f}")
print(f"Box Precision:  {results.box.p.mean():.3f}")
print(f"Box Recall:     {results.box.r.mean():.3f}")
```

```python
for i, name in results.names.items():
    print(f"[{name}] Box AP50={results.box.ap50[i]:.3f}, Box AP50-95={results.box.ap[i]:.3f}")
```

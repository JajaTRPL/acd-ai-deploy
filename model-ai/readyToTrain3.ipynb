```python
from ultralytics import YOLO

model = YOLO("yolo11n-seg.pt")

model.train(
    data="datasets2/data.yaml",
    epochs=200,
    imgsz=640,
    device=0,
    batch=16,
    patience=50,
    lr0=0.001,
    lrf=0.0001,
    momentum=0.95,
    weight_decay=0.0001,
    warmup_epochs=5,
    box=10.0,
    cls=0.8,
    dfl=2.0,
    mask_ratio=6,
    overlap_mask=True,
    augment=True,
)
```

```python
from ultralytics import YOLO

model = YOLO("runs/segment/train5/weights/best.pt")

results = model.val(
    data="datasets2/data.yaml",
    split="val",
    imgsz=640,
    batch=16,
    device=0
)

# ----- Overall metrics -----
print("Mask mAP50:     ", results.seg.map50)   # mAP@0.50 untuk mask
print("Mask mAP50-95:  ", results.seg.map)     # mAP@[0.50:0.95] untuk mask
print("Mask Precision: ", results.seg.p.mean())  # rata-rata P semua kelas
print("Mask Recall:    ", results.seg.r.mean())  # rata-rata R semua kelas)

print("Box mAP50:      ", results.box.map50)   # mAP@0.50 untuk bbox
print("Box mAP50-95:   ", results.box.map)     # mAP@[0.50:0.95] untuk bbox
print("Box Precision:  ", results.box.p.mean())
print("Box Recall:     ", results.box.r.mean())

for i, name in results.names.items():
    print(f"[{name}] Mask AP50={results.seg.ap50[i]:.3f}, Mask AP50-95={results.seg.ap[i]:.3f}")
```

```python
test_res = model.val(
  data="datasets2/data.yaml", 
  split="test", 
  imgsz=640, 
  batch=16, 
  device=0
  )

print("TEST Mask mAP50:", test_res.seg.map50, " | Mask mAP50-95:", test_res.seg.map)
print("TEST Box mAP50: ", test_res.box.map50,  " | Box mAP50-95: ", test_res.box.map)
```

Sweep conf/IoU

```python
from ultralytics import YOLO
import numpy as np

model = YOLO("runs/segment/train5/weights/best.pt")
bests = (0,0,0,0)  # (map, conf, iou, map50)

for conf in np.linspace(0.15, 0.65, 6):
    for iou in np.linspace(0.45, 0.75, 7):
        r = model.val(
            data="datasets2/data.yaml", 
            split="val",
            imgsz=640, 
            batch=16, 
            device=0,
            conf=float(conf), iou=float(iou)
        )
        if r.seg.map > bests[0]:
            bests = (float(r.seg.map), float(conf), float(iou), float(r.seg.map50))

print(f"Best mAP50-95={bests[0]:.3f} @ conf={bests[1]:.2f}, iou={bests[2]:.2f} (mAP50={bests[3]:.3f})")
```
